```python
data = pd.read_csv('data/insurance.csv')

# Definisikan nama label
sex = {
    'male': 1,
    'female': 0
}

smoker = {
    'yes': 1,
    'no': 0
}

region = {
    'northwest': 0,
    'southwest': 1,
    'southeast': 2,
    'northeast': 3
}

# Copy Data Frame untuk menyimpan Data Frame baru
# dengan label yang mudah untuk dibaca

# Ubah label dengan fungsi mapping dari Pandas
# pada Data Frame df_label
data['sex'] = data['sex'].map(sex)
data['smoker'] = data['smoker'].map(smoker)
data['region'] = data['region'].map(region)

# Cek Data Frame df_label
data.head()
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor

def calc_vif(X):

    # Calculating VIF
    vif = pd.DataFrame()
    vif["variables"] = X.columns
    vif["VIF"] = [variance_inflation_factor(
        X.values, i) for i in range(X.shape[1])]

    return(vif)
```

```python
dataX = data.iloc[:, :-1]
calc_vif(dataX)
```

```python
#Baris BMI dihapus 
dataX = data.drop(columns=['bmi'])
dataX.head()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn import metrics
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder

X_m = dataX.iloc[:, 0:4].values
y_m = dataX.iloc[:, -1].values

y_m = y_m.reshape(len(y_m), 1)


ct = ColumnTransformer(
    transformers=[('state', OneHotEncoder(), [3])], remainder='passthrough')
X_m = np.array(ct.fit_transform(X_m))

X_m_train, X_m_test, y_m_train, y_m_test = train_test_split(
    X_m, y_m, test_size=0.2, random_state=30)

regressor = LinearRegression()

regressor.fit(X_m_train, y_m_train)

y_m_pred = regressor.predict(X_m_test)

df = pd.DataFrame({'Actual': y_m_test.flatten(),
                   'Predicted': y_m_pred.flatten()})

print('R2 Score:', metrics.r2_score(y_m_test, y_m_pred))
```
